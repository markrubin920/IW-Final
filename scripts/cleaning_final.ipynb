```python
import pandas as pd
import numpy as np
import pybaseball
```

```python
df = pd.read_csv('../data/cleaned/clean_2020_2024.csv')
```

Initial Data 2020-2024

```python
df
```

```python
df.drop({"Unnamed: 0", "game_type"}, axis=1, inplace = True)

df
```

```python
df.columns
```

Encoding Outcomes

```python
set(df["description"])
```

```python
outcome_map = {'ball': 0,
 'blocked_ball': 0,
 'bunt_foul_tip': 3,
 'called_strike': 1,
 'foul': 3,
 'foul_bunt': 3,
 'foul_pitchout': 3,
 'foul_tip': 3,
 'hit_by_pitch': 0,
 'hit_into_play': 3,
 'missed_bunt': 2,
 'pitchout': 0,
 'swinging_strike': 2,
 'swinging_strike_blocked': 2}

df["Outcome"] = df["description"].map(outcome_map)

df
```

```python
df["pitch_name"].value_counts()
```

Reduce to only the 9 most common pitch types

```python
pitch_types = set(pd.DataFrame(df["pitch_name"].value_counts().head(9)).reset_index()["pitch_name"])
df = df[df["pitch_name"].isin(pitch_types)]
```

```python
df
```

Reduce to only top 50 players

```python
top_50 = df["batter"].value_counts().head(50).reset_index()

display(pd.DataFrame(top_50))
```

```python
def get_player_name(id):
    player_info = pybaseball.playerid_reverse_lookup([id])
    return player_info['name_first'][0] + " " + player_info['name_last'][0]

top_50["Player Name"] = top_50["batter"].map(get_player_name)
```

```python
top_50
```

```python
main_batters = set(top_50["batter"])
df = df[df["batter"].isin(main_batters)]

df
```

One Hot Encoding Pitches

```python
dummies = pd.get_dummies(df['pitch_name'], prefix='pitch_name')
df = pd.concat([df, dummies], axis=1)
```

```python
dummies_zone = pd.get_dummies(df['zone'], prefix='zone')
df = pd.concat([df, dummies_zone], axis=1)
```

```python
df
```

Encoding Handedness

```python
df.columns
```

```python
description_map = {'R': 0, 'L': 1}

df['Batter Side'] = df['stand'].map(description_map)
df['Pitcher Side'] = df['p_throws'].map(description_map)

df
```

Distributions by Pitch Type

```python
df.groupby('pitch_name').describe()["release_speed"]
```

```python
# Fixing lefty issues
df["vx0"] = abs(df["vx0"])
df["pfx_x"] = abs(df["pfx_x"])
df["ax"] = abs(df["ax"])
df["api_break_x_batter_in"] = abs(df["api_break_x_batter_in"])
```

```python
df.columns
```

Determining Feature vs Outcome

```python
features = [
    'release_speed', 'effective_speed', 'vx0', 'vy0', 'vz0', 'ax', 'ay', 'az', # Measures of speed and acceleration
    'release_pos_x', 'release_pos_z', 'release_extension', 'release_pos_y', # Release position
    'zone', # Pitch location
    'balls', 'strikes', 'on_3b', 'on_2b', 'on_1b', # Context
    'release_spin_rate', 'spin_axis', # Spin
    'api_break_z_with_gravity', 'api_break_x_arm', 'api_break_x_batter_in', 'pfx_x', 'pfx_z', # Break
    # Dummy variables for pitch type
    'pitch_name_4-Seam Fastball', 'pitch_name_Changeup',
    'pitch_name_Curveball', 'pitch_name_Cutter', 'pitch_name_Knuckle Curve',
    'pitch_name_Sinker', 'pitch_name_Slider', 'pitch_name_Split-Finger',
    'pitch_name_Sweeper', 'Batter Side', 'Pitcher Side'
    # 'outs_when_up', 'inning', 'batter_days_since_prev_game', 'pitch_number', 'bat_win_exp', 'n_priorpa_thisgame_player_at_bat', # Additional context that's probably not useful
    # 'plate_x', 'plate_z', 'sz_top', 'sz_bot', # Don't use for now but may be useful for making a distance for zone metric (redundant to zone)
]

outcomes = ['Outcome']
```

Making Dummy Variables for Players on Base

```python
df["on_1b"] = np.where(df["on_1b"] > 0, 1, 0)
df["on_2b"] = np.where(df["on_2b"] > 0, 1, 0)
df["on_3b"] = np.where(df["on_3b"] > 0, 1, 0)
```

```python
counts = pd.DataFrame(df.isnull().sum()).reset_index()
counts[counts[0] != 0]
```

```python
df = df[~df['effective_speed'].isna() & ~df['spin_axis'].isna() & ~df['release_extension'].isna() & ~df['api_break_z_with_gravity'].isna() & ~df['release_spin_rate'].isna() & ~df['pfx_x'].isna()]
```

```python
counts = pd.DataFrame(df.isnull().sum()).reset_index()
counts[counts[0] != 0]
```
