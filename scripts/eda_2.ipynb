```python
px.histogram(df_judge, 'pitch_name', 'estimated_ba_using_speedangle', histfunc="avg")
```

```python
df_subset = df_judge[df_judge["pitch_name"].isin(["4-Seam Fastball", "Slider", "Changeup"])]
df_subset["speed_bin"] = pd.cut(df_subset["release_speed"], bins=range(60, 105, 1))


df_grouped = (
    df_subset.groupby(["speed_bin", "pitch_name"], as_index=False)
    ["estimated_ba_using_speedangle"].mean()
)

df_grouped["speed_mid"] = df_grouped["speed_bin"].apply(lambda x: x.mid)

px.line(
    df_grouped,
    x="speed_mid",
    y="estimated_ba_using_speedangle",
    color="pitch_name",
    markers=True
)
```
